```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("Libraries imported successfully!")
```

```python
# Sample resumes
resumes = [
    "Python Machine Learning Data Analysis SQL Deep Learning",
    "Java Web Development HTML CSS JavaScript",
    "Python Data Science Machine Learning AI NLP",
    "Accounting Finance Excel Taxation"
]

# Candidate names
candidate_names = [
    "Candidate A",
    "Candidate B",
    "Candidate C",
    "Candidate D"
]

# Job description
job_description = "Python Machine Learning Data Science NLP SQL"

print("Resume data created successfully!")
```

```python
# Combine resumes and job description
all_text = resumes + [job_description]

# Convert text into numbers using TF-IDF
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(all_text)

print("Text converted successfully!")
```

```python
# Calculate similarity
similarity_scores = cosine_similarity(
    tfidf_matrix[-1],   # Job description
    tfidf_matrix[:-1]   # Resumes
)

# Convert scores into percentages
scores = similarity_scores[0] * 100

# Create results table
results = pd.DataFrame({
    "Candidate": candidate_names,
    "Match Score (%)": scores
})

# Sort candidates by highest score
results = results.sort_values(
    by="Match Score (%)",
    ascending=False
)

print(results)
```

```python
import matplotlib.pyplot as plt

# Plot candidate scores
plt.figure(figsize=(8,5))

plt.bar(
    results["Candidate"],
    results["Match Score (%)"]
)

plt.title("Resume Match Scores")
plt.xlabel("Candidates")
plt.ylabel("Match Score (%)")

plt.show()
```
